Combine DF's

```python
attendance = pd.concat([df2011, df2012, df2013, df2014, df2015, df2016, df2017]).reset_index(drop=True)
```

```python
attendance['year'].replace('2016-17',2017,inplace=True)
```

```python
attendance['k_8'] = attendance['k_8'].astype(float)
```

```python
attendance['9_12'] = attendance['9_12'].astype(float)
```

```python
attendance.info()
```

```python
x1 = attendance['year']
y1 = attendance['k_8']
attendance.plot(kind='line', x = 'year', y = 'k_8')
# Calculate the Trendline

z = np.polyfit(x1, y1, 1)
p = np.poly1d(z)
plt.plot(attendance['year'], p(x1), color="purple", linewidth=3, linestyle="--")
plt.title('Attendance Percent by Year')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.xticks(x1)
```

```python
x3 = attendance['year']
y3 = attendance['9_12']
attendance.plot(kind='line', x = 'year', y = '9_12')
# Calculate the Trendline

z = np.polyfit(x3, y3, 1)
p = np.poly1d(z)
plt.plot(attendance['year'], p(x3), color="purple", linewidth=3, linestyle="--")
plt.title('Attendance Percent by Year')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.xticks(x3)
```
